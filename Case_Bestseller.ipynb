Hipótese do cliente: livros de ficção eram preferidos pelos consumidores mas ultimamente pode estar acontecendo uma mudança de gosto

Para avaliar a hipótese do cliente, serão criadas visualizações que comparam parâmetro de preferência (rating médio e numero na lista) por genero ao longo dos anos

#Rating médio dos Best Seller por Genero e por ano

```python
year_rating_mean  = pd.crosstab(data_set.Year, data_set.Genre, values = data_set[['User Rating']], aggfunc='mean').round(3) 
year_rating_mean
```

##Vizualização
Observa-se que em geral livros de ficção tem um rating médio maior



```python
yf = year_rating_mean.iloc[:, 0]
ynf = year_rating_mean.iloc[:, 1]
x = pd.DataFrame(year_rating_mean.index)
plt.title('Rating médio por genero')
plt.xlabel ('Anos')
plt.ylabel ('Rating médio')
plt.grid(axis = 'both')
plt.plot(x, yf, label = 'Fiction' )
plt.plot(x, ynf, label = 'Non Fiction')
plt.legend()
```

#Best_Sellers anualmente por genero



```python
year_genre  = pd.crosstab(index = data_set['Year'], columns = data_set['Genre'])
year_genre
```

##Vizualização
### É possivel observar que ao longo dos ultimos anos houve uma preferência por livros de Não Ficção, exceto no ano de 2014

```python
yf = pd.DataFrame(year_genre.iloc[:, 0])
ynf = pd.DataFrame(year_genre.iloc[:, 1])
xYear = pd.DataFrame(year_genre.index)

plt.title ('Genero dos Best Sellers')
plt.xlabel ('Anos')
plt.ylabel('Numero de livros')
plt.grid(axis = 'both')
plt.plot(xYear, yf, label = 'Fiction', linestyle = "solid")
plt.plot(xYear, ynf, label = 'Non Fiction', linestyle = "solid") 
plt.legend(title = 'Genre', loc=3)
```

#Média de preço dos Best Sellers por genero

```python
price_mean = pd.crosstab(data_set.Year, data_set.Genre, values= data_set.Price, aggfunc='mean')
price_mean
```

## Vizualização
Talvez exista alguma correlação entre o preço médio do livro por genero e o numero de best sellers

```python
yf = price_mean.iloc[:, 0]
ynf = price_mean.iloc[:, 1]
x = pd.DataFrame(price_mean.index)
plt.title('Preço médio por genero')
plt.xlabel ('Anos')
plt.ylabel ('Preço médio')
plt.grid(axis = 'both')
plt.plot(x, yf, label = 'Fiction' )
plt.plot(x, ynf, label = 'Non Fiction')
plt.legend()
```

#Resumo da análise
Apesar dos livros de ficção terem maior rating médio, nos ultimos anos, livros de não ficção tiveram maior presença entre os mais vendidos. Além disso, livros de não ficção apresentam maior preço médio. Portanto a mudança de foco para vender e estocar livros de não ficção pode atrair mais clientes e existe um potencial de aumentar a receita. Aumento da receita representa uma oportunidade de aumento de lucro quando assiciado à manutenção dos custos




Outra hipótese trazida pelos donos da empresa é a diversificação de livros na lista de bestsellers ao longo dos anos, de forma que os livros estariam ficando menos tempo na lista de bestsellers com o passar do tempo. 


Criando uma tabela para associar o ano de entrada do livro na lista dos best sellers e o tempo de permanência

```python
a = pd.crosstab(data_set.Name, data_set.Year, values= data_set.Year, aggfunc='count')
a.fillna(value=0, inplace= True) 

a['Years_inlist'] = a.sum(axis=1)
a = a.reset_index()
a['Start'] = 0
```

```python
for i in range(len(a)):
  for j in range(len(a.columns)):
    if (a.iloc[i, j] == 1 and a.iloc[i, -1] == 0):
      a.iloc[i, -1] = j + 2008

a
```

```python
b = pd.crosstab(a.Years_inlist, a.Start, values= a.Start, aggfunc='count')
b.fillna(value = 0, inplace = True)
b = b.reset_index()
b
```

```python
c = b.iloc[0, 1:]
plt.title('Livros entrantes na lista')
plt.xlabel ('Anos')
plt.ylabel ('Novos livros na lista')
plt.grid(axis = 'both')
plt.plot(c)
```

Para algum livro novo entrar, outro que ja estava na lista tem que sair. Observa-se que no periodo estudado, no primeiro ano houve uma entrada grande de livros novos seguido por uma queda que se manteve ao longo dos anos. Isso pode ser um indicativo de que a hipótese de que os livros estão passando menos tempo é falsa. 

Por fim, a alta direção acredita que as notas médias dos livros possuem maior influência na quantidade de vendas e tempo na lista de bestsellers do que os seus preços.
